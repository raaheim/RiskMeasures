##### 2) Simulate $M$ paths of the stock $S$

```python
B = np.random.binomial(1,p,T+1) # We simulate all the Bernouilli random variables
```

```python
def transform(B): # Transform a vector of O and 1 into a vector of u and d's
    temp1 = np.size(B)
    up_and_down = np.zeros(temp1)
    for i in range(0,temp1):
        if B[i]==1:
            up_and_down[i]=u
        else:
            up_and_down[i]=d
    return up_and_down
```

```python
def simulateS(m):
    B = np.zeros(m)
    up_and_down = np.zeros(m)
    for i in range(1,T+1):
        B = np.random.binomial(1,p,m)
        up_and_down = transform(B)
        S[:,i]=S[:,i-1]*up_and_down
```

### Another representation using graphs (Binary trees)

```python
def binomial_grid(n,s0):
    G=nx.Graph() 
    for i in range(0,n):
        j=-i+1
        while (j<i+2):
            G.add_edge((i,j),(i+1,j+1),weight=0.0)
            G.add_edge((i,j),(i+1,j-1),weight=0.0)
            j=j+2
    
    j=1
    for i in range(0,n):
        r=np.random.binomial(1,p,1)
        if r >0:
            G.add_edge((i,j),(i+1,j+1),weight=1.0)
            j=j+1
        else:
            G.add_edge((i,j),(i+1,j-1),weight=1.0)
            j=j-1
    
    posG={}
    lab={}
    for node in G.nodes():
        posG[node]=(node[0],node[1])
        if node[0]==0:
            lab[node]=s0
        i=node[0]
        j=node[1]
        if j>1:
            lab[node]=s0*(u**(j-1))
        elif j==1:
            lab[node]=s0
        else:
            lab[node]=s0*(d**(1-j))
    elarge=[(a,b) for (a,b,c) in G.edges(data=True) if c['weight'] >0.5]
    esmall=[(a,b) for (a,b,c) in G.edges(data=True) if c['weight'] <=0.5]
    fig, ax = plt.subplots(1,1,figsize = (12,8))
    nx.draw_networkx_edges(G,posG,edge_color='blue',width=1)
    nx.draw_networkx_labels(G,posG,lab,font_size=15,font_family='sans-serif')
    plt.xlim(-1, n+1)
    plt.ylim(-n-1, n+2)
    plt.yticks([], [])
    plt.xlabel("time", size = 20)
    plt.legend(["Value of risky asset"], loc = 2, fontsize = 12)
    plt.show()
```

### Pricing

##### Pricing of European Financial claims on the Cox-Ross-Rubinstien model

The aim of this section is to write a function "Price" which computes the (arbitrage) price at any time of a contract whose payoff is given by a mapping $g_T$ using the Dynamic Programming Principle as in Theorem 6.2.1.

we start with a plotting function for a Binary tree

```python
def plot_tree(g):
    pos={}
    lab={}
    
    for n in g.nodes():
        pos[n]=(n[0],n[1])
        lab[n]=float(int(g.node[n]['value']*1000))/1000 # This is just to print only with 10^-2 precision
    
    elarge=g.edges(data=True)
    nx.draw_networkx_edges(g,pos,edgelist=elarge)
    nx.draw_networkx_labels(g,pos,lab,font_size=15,font_family='sans-serif')
    plt.xlim(-1, T+1)
    plt.ylim(-T-1, T+2)
    plt.show()   
```

We construct the graph of prices

```python
def graph_stock():
    S=nx.Graph()
    for i in range(0,T):
        j=-i+1
        while (j<i+2):
            S.add_edge((i,j),(i+1,j+1))
            S.add_edge((i,j),(i+1,j-1))
            j=j+2
            
    for n in S.nodes():
        
        if n[0]==0:
            S.node[n]['value']=S0
        i=n[0]
        j=n[1]
        if j>1:
            S.node[n]['value'] = S0*(u**(j-1))
        elif j==1:
            S.node[n]['value'] = S0
        else:
            S.node[n]['value'] = S0*(d**(1-j))
    return S
```

##### 3) Compute the price of a European Call Option

```python
def European_call(x,K):
        return np.maximum(x-K,0.0)
```

```python
def compute_European_Call_prices(K):

    g = nx.Graph()
    price = nx.Graph()
    
    S = nx.Graph()
    
    S = graph_stock()

    for i in range(0,T):
            j=-i+1
            while (j<i+2):
                g.add_edge((i,j),(i+1,j+1))
                g.add_edge((i,j),(i+1,j-1))
                price.add_edge((i,j),(i+1,j+1))
                price.add_edge((i,j),(i+1,j-1))
                j=j+2

    for n in S.nodes():
        if n[0]==T:
            g.node[n]['value'] = European_call(S.node[n]['value'],K)
            price.node[n]['value'] = European_call(S.node[n]['value'],K)
    
    i=T-1
    while i>=0:
        j=-i+1
        while j<i+2:
            g.node[(i,j)]['value'] = g.node[(i+1,j+1)]['value']*p_star + g.node[(i+1,j-1)]['value']*(1.0-p_star)
            price.node[(i,j)]['value'] = g.node[(i,j)]['value']*((1.0+r)**(i-T))
            j=j+2
        i=i-1
        
    return price
```

Computation the price of a European Call option

##### 4) Compute the price of a European Put Option

```python
def compute_European_Put_prices(K):

    Put = nx.Graph()
    
    S = nx.Graph()
    S = graph_stock()
    
    Call = nx.Graph()
    Call = compute_European_Call_prices(K)

    for i in range(0,T):
            j=-i+1
            while (j<i+2):
                Put.add_edge((i,j),(i+1,j+1))
                Put.add_edge((i,j),(i+1,j-1))
                j=j+2
    
    for n in S.nodes():
        if n[0]==T:
            Put.node[n]['value'] = European_call(K,S.node[n]['value'])    
    i=T-1
    while i>=0:
        j=-i+1
        while j<i+2:
            Put.node[(i,j)]['value'] = Call.node[(i,j)]['value'] + K*(1.0+r)**(i-T) - S.node[(i,j)]['value']
            j=j+2
        i=i-1
        
    return Put
```

##### 5) Complete the codes above to compute a hedging strategy and simumate the associated wealth processes

```python
def compute_European_Call_strategy(K):

    Call = nx.Graph()
    Call = compute_European_Call_prices(float(K))

    S = nx.Graph()
    S = graph_stock()
    
    Strat_Call = nx.Graph()
    
    for i in range(0,T):
            j=-i+1
            while (j<i+2):
                Strat_Call.add_edge((i,j),(i+1,j+1))
                Strat_Call.add_edge((i,j),(i+1,j-1))
                j=j+2
    
    for n in S.nodes():
        if n[0]==T:
            Strat_Call.node[n]['value'] = 0.0    
    i=T-1
    while i>=0:
        j=-i+1
        while j<i+2:
            Strat_Call.node[(i,j)]['value'] = (Call.node[(i+1,j+1)]['value']-Call.node[(i+1,j-1)]['value'])/(S.node[(i,j)]['value']*(u-d))
            j=j+2
        i=i-1
        
    return Strat_Call
```

```python
def compute_European_Put_strategy(K):

    Put = nx.Graph()
    Put = compute_European_Put_prices(float(K))

    S = nx.Graph()
    S = graph_stock()
    
    Strat_Put = nx.Graph()
    
    for i in range(0,T):
            j=-i+1
            while (j<i+2):
                Strat_Put.add_edge((i,j),(i+1,j+1))
                Strat_Put.add_edge((i,j),(i+1,j-1))
                j=j+2
    
    for n in S.nodes():
        if n[0]==T:
            Strat_Put.node[n]['value'] = 0.0    
    i=T-1
    while i>=0:
        j=-i+1
        while j<i+2:
            Strat_Put.node[(i,j)]['value'] = (Put.node[(i+1,j+1)]['value']-Put.node[(i+1,j-1)]['value'])/(S.node[(i,j)]['value']*(u-d))
            j=j+2
        i=i-1
        
    return Strat_Put
```

```python
def compute_European_Call_optimal_Wealth(K):

    Call = nx.Graph()
    Call = compute_European_Call_prices(float(K))

    S = nx.Graph()
    S = graph_stock()
    
    Strat_Call = nx.Graph()
    Strat_Call = compute_European_Call_strategy(float(K))
    
    Wealth = nx.Graph()
    
    #Initialization of the tree wealth
    Wealth = Call 
    
    for i in range(0,T):
        j=-i+1
        while (j<i+2):
            Wealth.add_edge((i,j),(i+1,j+1))
            Wealth.node[(i+1,j+1)]['value'] = Strat_Call.node[(i,j)]['value']*S.node[(i+1,j+1)]['value'] + (Wealth.node[(i,j)]['value'] - Strat_Call.node[(i,j)]['value']*S.node[(i,j)]['value'])*(1+r) 

            Wealth.add_edge((i,j),(i+1,j-1))
            Wealth.node[(i+1,j-1)]['value'] = Strat_Call.node[(i,j)]['value']*S.node[(i+1,j-1)]['value'] + (Wealth.node[(i,j)]['value'] - Strat_Call.node[(i,j)]['value']*S.node[(i,j)]['value'])*(1+r) 
            j=j+2
        
    return Wealth
```

##### 6) We now deal with the case of American Options

We admit that the price at a given time $t$ of an American Option with respect to the family of discounted payoffs $(\tilde G_t:=(1+r)^{-t} g_t^{Europ}(S_t))_t$ is given by:
$$ \nu_t = \max\left(\tilde G_t,\mathbb{E}^{\mathbb{P}^\ast}[\nu_{t+1}\vert \mathcal{F}_t]\right), \quad \nu_T:=\tilde G_T. $$

Deduce that at any time $t$, $\nu_t = f_t^{Amer}(S_t)$ with
$$ f_t^{Amer}(x):= \max\left((1+r)^{-t} g_t^{Europ}(x),f_{t+1}^{Amer}(x u) p^\ast + f_{t+1}^{Amer}(x d) (1-p^\ast)\right), \quad f_T^{Amer}(x):=g_T^{Europ}(x).$$

Write a code to compute the graph of an American Call Option and compare it with the price of the European Call Option

```python
def compute_American_Call_Snell(K):
    
    Snell = nx.Graph()
    
    S = nx.Graph()
    
    S = graph_stock()
    
    for i in range(0,T):
        j=-i+1
        while (j<i+2):
            Snell.add_edge((i,j),(i+1,j+1))
            Snell.add_edge((i,j),(i+1,j-1))
            j=j+2
            
    for n in S.nodes():
        if n[0]==T:
            Snell.node[n]['value'] = European_call(S.node[n]['value'],float(K))*(1.0+r)**(-T) 
    
    i=T-1
    while i>=0:
        j=-i+1
        while j<i+2:
            temp = Snell.node[(i+1,j+1)]['value']*p_star + Snell.node[(i+1,j-1)]['value']*(1.0-p_star)
            Snell.node[(i,j)]['value'] = np.maximum(European_call(S.node[(i,j)]['value'],float(K))*(1.0+r)**(-i),temp)
            j=j+2
        i=i-1
        
    return Snell
```
